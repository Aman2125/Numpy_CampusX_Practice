```python
import numpy as np

a = np.array([1,2,3])
print(a) #1D array
```

```python
#2D and 3D array
b = np.array([[1,2,3] , [4,5,6]])
print(b)
```

```python
#3D also called tenser
c = np.array([[1,2] , [3,4] , [5,6]])
print(c)
```

```python
#dtype that changine int to other variables
a=np.array([1,2,3] , dtype=bool)
b=np.array([1,2,3] , dtype=float)
c=np.array([1,2,3] , dtype=complex)

print(a)
print(b)
print(c)
```

```python
#np.arange , also method for creating array , like range in python
a = np.arange(1 ,11)
b = np.arange(1,11,2) #stride = 2

print(a)
print(b)
```

```python
# with reshape , converts 1D to 2D
a = np.arange(1 ,11).reshape(5,2)
b = np.arange(1 ,11).reshape(2,5);

print(a)
print(b)
```

```python
# np.ones and np.zeros , all items 1 or 0 , used for initialization
a = np.ones((3,4)) #3rows and 4 col with 1,to initialize weights in NN
b = np.zeros((3,4))

print(a)
print(b)
```

```python
# np.random , initialise numbers between 0 and 1 randomly
np.random.random((3,4))
```

#Array Function

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
#max/min/sum/prod
np.sum(a1)
np.max(a1)
np.min(a1)
np.prod(a1)
np.max(a1 , axis=1) #for finding each row maxi
np.max(a1 , axis=0) #for finding each col maxi
```

```python
#mean/median/std/var

np.mean(a1 , axis=1) #for each row mean
np.var(a1)
```

```python
#trignometric functions
np.sin(a1)
```

```python
#dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
np.dot(a2 , a3)
```

```python
#log and exponents
np.exp(a1)
np.log(a1)
```

```python
#round/ceil/floor

a = np.random.random((2,3))*100
np.round(a)
np.floor(a)
np.ceil(a)
```

#Stacking

```python
#horizontal stacking means just piling up , shape must be same
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
np.hstack((a4,a5))
```

```python
#vertical stacking
np.vstack((a4,a5))
```
