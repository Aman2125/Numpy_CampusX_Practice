*   loc- searches using index labels



```python
#single row
movies.iloc[0] #indec position
```

```python
#multiple row
movies.iloc[0:5]
```

```python
#fancy indexing
movies.iloc[[0,4,5]]
```

```python
#loc
student
```

```python
student.loc['Aman'] # searching using index label here by using loc
```

#Filtering a DataFrame

```python
ipl.head(2)
```

```python
#find all the final winners
result = ipl['MatchNumber']=='Final'
new_df = ipl[result]
new_df[['Season','WinningTeam']]
```

```python
#how many super over finishes have occured
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
#how many matches has csk won in kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
x = ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]
y = ipl.shape[0]
ans=(x/y)*100
ans
```
