```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
```

```python
movies.head(1)
```

```python
#in groupby we form group on the basis of some col in the data
#group by always applies on categorical col and not on numerical one
# que form group on basis of genre
genres = movies.groupby('Genre')
genres
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.mean(numeric_only=True)
```

```python
# find top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
#or movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
#2nd way is faster as sum applied late
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#find the highest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
# find no of movies done by each actor
#movies['Star1'].value_counts()
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# Groupby attributes and methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```

```python
# find how many group formed when using groupby
len(movies.groupby('Genre'))
```

```python
# items in each size
movies.groupby('Genre').size()
```

```python
# first()/last() -> nth item
genres = movies.groupby('Genre')
#genres.first()
#genres.last()  # give last movie of each genre
genres.nth(4) #gives 5th movie of each genre
```

```python
#get_group -> vs filtering
genres.get_group('Horror') # horror movies
#or like old method
# movies[movies['Genre'] == 'Horror']
```

```python
# groups attribute
genres.groups # index postion of all the movie with that genre , gives dictionary
```

```python
# describe
genres.describe() # har group ke mathematical col ko summarise kr rha
```

```python
# sample
genres.sample() # one movie on each genre randomly
```

```python
# nunique
genres.nunique() #give unique movie on each genre
```

```python
# aggregate methods , we pass dict using agg

genres.agg(
    {
        'Runtime':'mean',
        'Gross':'sum',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
numeric_cols = movies.select_dtypes(include=np.number).columns.tolist()
genres[numeric_cols].agg(['min', 'max', 'mean'])
```

```python
# adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'Gross':['sum','max'],
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups

# Create an empty list to store DataFrames
df_list = []

for group,data in genres:
 # Append the highest rated movie for the current genre to the list
 df_list.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

# Concatenate all DataFrames in the list into a single DataFrame
df = pd.concat(df_list, ignore_index=True)
df
```

```python
# split (apply) combine
# apply -> builtin  function
genres.apply(min) #gives minimum value in each genre
```

```python
# find number of movies starting with A for each group

def fun(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(fun) # custom fuction passed
```

```python
# find rating of each movie in the group according to IMDB score

def ranking_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)
  return group
```

```python
genres.apply(ranking_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
# group by on multiple col
duo = movies.groupby(['Director' , 'Star1'])
duo

#size
duo.size()
```

```python
# get_group
duo.get_group(('Aamir Khan' , 'Amole Gupte'))
```

```python
# find the most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best (in term of metascore(avg)) actor -> genre combo
movies.groupby(['Star1' , 'Genre'])['Metascore'].mean().reset_index().sort_values('Metascore' ,ascending=False).head(1)
```

```python
# agg on multiple groupby
numeric_cols = movies.select_dtypes(include=np.number).columns.tolist()
duo[numeric_cols].agg(['min' , 'max' ,'mean']).head(2)
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl.sample(1)
```

```python
ipl.shape # very big data
```

```python
# find the top 10 batsman in terms of run scored
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)
```

```python
# find batsman with the most number of 4s and 6s in the last 5 overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)
```

```python
# find R sharma records against all teams
temp_df = ipl[ipl['batsman'] == 'RG Sharma']
temp_df.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# create a function that can return the highest score of any batsman
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('RG Sharma')
```
