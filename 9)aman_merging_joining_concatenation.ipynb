```python
import pandas as pd
import numpy as np
```

```python
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')
matches = pd.read_csv('matches.csv')
delivery = pd.read_csv('/content/deliveries (1).csv')
```

```python
# pd.concat , method to vertically stack two or more dataframe
regs = pd.concat([nov,dec] , ignore_index=True) # ignore_index for new index assignment and ignoring the previous index
regs.head(2)
```

```python
# multiindex dataframe , contains more than 1 index
multi = pd.concat([nov,dec],keys= ['Nov' , 'Dec'])
multi.loc['Nov']
# for extacting 1 element
# multi.loc[('Dec' , 4) ] , 4 is the index
```

```python
# for horizontal stacking use axis =1
pd.concat([nov,dec] , axis =1).head() # col must be same
```

#Merge

```python
#inner join , same like sql ,  joining two dataframe on the basis of one common col
# inner join is basically intersction , finding items which is common in both datafame
# que perfrom inner join on students and regs
students.merge(regs , how = 'inner' , on='student_id') #student left table , regs right table
# performing inner join on studentid col
```

```python
# left join , complete left table and common from right table
courses.merge(regs , how='left' , on='course_id')
```

```python
# right join
temp_df = pd.DataFrame({
    'student_id' : [26,27,28],
    'name': ['Aman','Jago','Mahi'],
    'partner' : [28,26,17]
})

students = pd.concat([students,temp_df] , ignore_index=True)
```

```python
students.tail()
```

```python
# here aman , jago , mahi have not enrolled in any of the courses but by performing right join we will get these result

students.merge(regs , how='right' , on='student_id')
```

```python
# full join or outer join
students.merge(regs , how='outer' , on='student_id').tail(10)
```

```python
# find total revenue generated by the company
total = regs.merge(courses,how='inner' , on='course_id')['price'].sum()
total
```

```python
# find month by month revenue
nov_rev = nov.merge(courses , how='inner' , on='course_id')['price'].sum()
print(nov_rev)
dec_rev = dec.merge(courses , how='inner' , on='course_id')['price'].sum()
print(dec_rev)
#or
```

```python
#2nd method
temp_df = pd.concat([nov,dec] , keys = ['Nov' , 'Dec']).reset_index()
temp_df.merge(courses , on='course_id').groupby('level_0').sum()
```

```python
# print the registration table
# cols -> name -> course -> price
regs.merge(students , on='student_id').merge(courses , on='course_id')[['name' , 'course_name' , 'price']].head()
```

```python
# plot a bar chart for revenue/course
regs.merge(courses , on='course_id').groupby('course_name').sum().plot(kind='bar')
```

```python
# find students who enrolled in both the months
common_stu = np.intersect1d(nov['student_id'] , dec['student_id'])
common_stu
```

```python
students[students['student_id'].isin(common_stu)]
```

```python
# find course that got no enrollment
course_id_list = np.setdiff1d(courses['course_id'] , regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
# find students who did not enroll into any courses
student_id_list = np.setdiff1d(students['student_id'] , regs['student_id'])
students[students['student_id'].isin(student_id_list)]
```

```python
# print student name -> partner name for all enrolled students
# self join here
students.merge(students , how='inner' , left_on= 'partner' , right_on='student_id')[['name_x' , 'name_y']]
```

```python
# find top 3 students who did most number enrollments
regs.merge(students , on='student_id').groupby(['student_id' ,'name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# find top 3 students who spent most amount of money on courses
regs.merge(students , on='student_id').merge(courses , on='course_id').groupby(['student_id' , 'name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# Alternate syntax for merge
pd.merge(students,regs,how='inner' , on='student_id')
```
